```python
import tensorflow as tf
import keras_preprocessing
from keras_preprocessing import image
from keras_preprocessing.image import ImageDataGenerator
```

```python
TRAINING_DIR = "dataset2/train"
training_datagen = ImageDataGenerator(rescale = 1./255,
                                      rotation_range=40,
                                      width_shift_range=0.2,
                                      height_shift_range=0.2,
                                      shear_range=0.2,
                                      zoom_range=0.2,
                                      horizontal_flip=True,
                                      fill_mode='nearest')

train_generator = training_datagen.flow_from_directory(TRAINING_DIR,
                                                        target_size=(128,128),
                                                        class_mode='binary',
                                                        batch_size=90)
```

```python
VALIDATION_DIR = "dataset2/valid"
validation_datagen = ImageDataGenerator(rescale = 1./255)

validation_generator = validation_datagen.flow_from_directory(VALIDATION_DIR,
                                                              target_size=(128,128),
                                                              class_mode='binary',
                                                              batch_size=15)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(8, (3,3), activation='relu', input_shape=(128, 128, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(16, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])


model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
import numpy as np
import pandas as pd
from keras.preprocessing import image
import os
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

images = os.listdir("dataset2/test/yes/")
k = len(images)
print(images)
count_total=0
count_yes=0
count_no = 0
dict = {'Img_name':[],'Actual':[],'prediction':[]}
df_yes = pd.DataFrame(dict) 
for i in images:
    path = "dataset2/test/yes/" + i
    img = image.load_img(path, target_size=(128,128))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)

    images = np.vstack([x])
    classes = model.predict(images, batch_size=10)
    count_total=count_total+1
    print(count_total," done out of",k)
    if(classes[0][0]==1):
        count_yes = count_yes+1
        df2 = {'Img_name':i,'Actual':'yes','prediction':'yes'}
        df_yes = df_yes.append(df2, ignore_index = True)
    else:
        count_no = count_no+1
        df2 = {'Img_name':i,'Actual':'yes','prediction':'no'}
        df_yes = df_yes.append(df2, ignore_index = True)
    
```
